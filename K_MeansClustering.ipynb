# K-Means

K-Means algorithm is a centroid based clustering algorithm where the sum of distances of points from the centroid of each cluster is minimized. The final output is a set of K clusters .ie. the cluster assigned to each point and the K centroids of the clusters.

### The Algorithm
> 1.   Initialize K centroids to K points randomly and set each point's initial cluster as the centroid closest to it.
2.   While the clusers are changing - \\
a. Assign the new centroids as the centroids of the points which have the same assigned cluster. \\
b. Assign the new clusters to the points as the closest centroid.
3. Return the assignments and the centroids.





### Part 1 - Implementation of KMeans and testing on Synthetic Datasets

```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating Synthetic Data

k=2
color=["darkturquoise","darkorange","teal","darkviolet","tomato", "yellowgreen","hotpink","gold"]

X,_=make_blobs(n_samples=500, n_features=2, centers=k, random_state=10)
print("Shape = " + str(X.shape))

print("Sample:")
print(X[:7])
```

####Define all the functions & varibles needed to do clustering

```python
#Create an empty dictionary to store the cluster name, centroid and points associated with the cluster.
clusters={}

def initializeClusterCentroids():
  #For each cluster, initialize its properties
  for i in range(k):
      center = np.random.randint(-8, 10, size= (X.shape[1],), )
      points = []

      cluster = {
          'center' : center,
          'points' : points,
          'color' : color[i]
      }
      clusters[i] = cluster #looping over clusters dictionary

#Calculate euclidean distance between v1 and v2
def distance(v1, v2):
    return np.sum((v2-v1)**2)**0.5

def assignPointsToCluster(clusters, X):
    for ix in range(X.shape[0]):#iterate for all datapoints

        #Foe each datapoint, find the distance to k centroids
        distance_of_i = []
        for kx in range(k):
            d = distance(X[ix], clusters[kx]['center']) #we have find distance b/w
            #all the point with all the five (k) cluster centers
            distance_of_i.append(d)

        #Based on the distances to k centroid, pick which cluster to assign it to
        #here we want find the minimum distance of cluster centers
        cluster_to_choose = np.argmin(distance_of_i)
        clusters[cluster_to_choose]['points'].append(X[ix])

#STEP 3: UPDATE CLUSTER CENTROIDS
def updateCluster(clusters):
    for kx in range(k):
        cluster_points = clusters[kx]['points']

        cluster_points = np.array(cluster_points)
        #here pts are array of list but we want array of array so

        if len(cluster_points)>0:
            new_center = np.mean(cluster_points, axis=0) #axis is along rows so will
            #find mean of all the feature seprately it will give 2,0 when we have cluster pts 50,2
            clusters[kx]['center'] = new_center
            clusters[kx]['points'] = [] # clear the points in a cluster list (emptying pts)
            #as we have to do step 2 after step 3 again till not converge
```

```python
#We need to visualize this too. Lets define a function that can do that!
def plotclusters(clusters, plot):
    for kx in range(k):
        cluster_points = clusters[kx]['points']
        cluster_color = clusters[kx]['color']
        cluster_center = clusters[kx]['center']
        cluster_points = np.array(cluster_points)

        #plotting points associated /nearest to the cluster centers
        if len(cluster_points) > 0:
          plot.scatter(cluster_points[:, 0], cluster_points[:, 1], s = 2, c = cluster_color)

        plot.scatter(cluster_center[0], cluster_center[1], s = 250, c = cluster_color, marker="o")
```

We can now run and visualize how the clusters evolve.

```python
#Plotting the initial dataset
fig, axs = plt.subplots(2, 3)
fig.set_size_inches(15, 10)
axs[0, 0].scatter(X[:,0],X[:,1], s = 2, c = 'black')
axs[0, 0].set_title('Dataset')

np.random.seed(1)
initializeClusterCentroids()

#Now, lets plot the centroids it has initalized
axs[0, 1].set_title('Centroid Initialization')
axs[0, 1].scatter(X[:, 0], X[:, 1], s = 2, c = 'black')
for i in range(k):#for all the clusters
    center = clusters[i]['center'] #its cluster's center
    axs[0, 1].scatter(center[0], center[1], c=clusters[i]['color'], s=250, marker="o")

#CLUSTERING EPOCH 1
axs[0, 2].set_title('Iteration 1')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[0, 2])
updateCluster(clusters)

#CLUSTERING EPOCH 2
axs[1, 0].set_title('Iteration 2')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 0])
updateCluster(clusters)

#CLUSTERING EPOCH 3
axs[1, 1].set_title('Iteration 3')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 1])
updateCluster(clusters)

#CLUSTERING EPOCH 4
axs[1, 2].set_title('Iteration 4')
assignPointsToCluster(clusters, X)
plotclusters(clusters, axs[1, 2])
updateCluster(clusters)

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### KMeans on Digits Dataset (KMeans for labelling)

Here we will attempt to use k-means to try to identify similar digits without using the original label information; this might be similar to a first step in extracting meaning from a new dataset about which you don't have any a priori label information.

```python
# importing the dataset
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
```

```python
# importing Kmeans from scikit-learn
from sklearn.cluster import KMeans

# Perform KMeans
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
# We can now see how the centers of the clusters formed are

fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

Because k-means knows nothing about the identity of the cluster, the 09 labels may be permuted. We can fix this by matching each learned cluster label with the true labels found in them:

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

Now we can check how accurate our unsupervised clustering was in finding similar digits within the data:

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

**Hence clustering algorithms like KMeans are highly effective for identifying labels for groups apriori and even a simple KMeans acheives a ~80% accuracy on digit classification.**

The confusion matrix for this looks like:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()

mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

### KMeans on Iris Dataset (Elbow Method)

We will perform KMeans on the famous Iris Dataset which has four features: sepal length, sepal width, petal length, and petal width. The fifth column is for species, which holds the value for these types of plants.

```python
#Importing the neccesary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#import the dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
x = df.iloc[:, [0,1,2,3]].values
df.head(5)
```

Let us start with assigning the initial number of clusters as k=5.

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
```

How do we choose the optimal number of clusters? Theres a method called the Elbow method, which is designed to help find the optimal number of clusters in a dataset. So lets use this method to calculate the optimum value of k. To implement the Elbow method, we need to create some Python code (shown below), and well plot a graph between the number of clusters and the corresponding error value.

This graph generally ends up shaped like an elbow, hence its name:

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)

plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

The output graph of the Elbow method shows that the shape of elbow is approximately formed at k=3. The optimal value of k is between 2 and 4, as the elbow-like shape is formed at k=3 in the above graph.

```python
kmeans3 = KMeans(n_clusters=3)
y_kmeans3 = kmeans3.fit_predict(x)
print(y_kmeans3)

kmeans3.cluster_centers_
```

We can now visualize the clustering in two dimensions instead of the original four.

```python
plt.scatter(x[:,0], x[:,1], c=y_kmeans3, cmap='rainbow')
```

ANSWER1

Feature scaling is a data preprocessing step that aims to normalize the range of features in a dataset. It is essential for KMeans as it is for most ML algorithms, because it affects the distance calculation between data points and cluster centers. If the features have different scales, the algorithm may assign more weight to the features with higher magnitudes and create biased clusters.

To perform feature scaling, you can use different methods such as min-max normalization, standardization, or robust scaling. In Python, you can use the StandardScaler, MinMaxScaler, or RobustScaler classes from the sklearn.preprocessing module to scale your data before applying KMeans. You can also use the scale function to convert your data to z-scores, as suggested in this tutorial1.

Here is an example of how to scale your data using StandardScaler and apply KMeans in Python:

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generate some random data
np.random.seed(42)
x = np.random.randn(1000) / 2
y = np.concatenate([np.random.randn(500) + 5, np.random.randn(500)]) / 10

# Plot the data without scaling
plt.scatter(x, y, s=2)
plt.title("Data without scaling")
plt.show()

# Scale the data using StandardScaler
scaler = StandardScaler()
X = np.array([x, y]).T
X_scaled = scaler.fit_transform(X)

# Plot the scaled data
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=2)
plt.title("Data with scaling")
plt.show()

# Apply KMeans with 2 clusters
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(X_scaled)

# Plot the clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap="bwr", s=2)
plt.title("KMeans clusters")
plt.show()
```

Initialization variation in KMeans refers to the problem of getting different clustering results depending on the initial choice of centroids. This can affect the quality and stability of the algorithm, as well as the number of iterations required to converge.

There are several ways to prevent or reduce initialization variation in KMeans, such as:

Using a deterministic method to select the initial centroids, rather than a random one. For example, you can use the KMeans++ algorithm1, which chooses the initial centroids based on the distance between data points, ensuring that they are spread out and close to the true clusters. Another example is the Kaufmans method2, which chooses the initial centroids based on the density of data points, ensuring that they are representative of the data distribution.
Using a robust method to select the initial centroids, rather than a sensitive one. For example, you can use the Wards method3, which is a hierarchical clustering technique that minimizes the variance within each cluster, ensuring that the initial centroids are compact and coherent. Another example is the median method4, which chooses the initial centroids as the median of each feature, ensuring that they are resistant to outliers and noise.
Running the KMeans algorithm multiple times with different random initializations, and choosing the best solution based on some criterion, such as the within-cluster sum of squares (WSS) or the silhouette score. This way, you can increase the probability of finding the global optimum, rather than a local one, and compare the results across different runs to assess the stability of the algorithm.

The training and testing complexity of KMeans depends on the algorithm used to find the optimal cluster centers. The most common algorithm is Lloyds algorithm, which iterates over the data points and assigns them to the nearest cluster center, then updates the cluster centers as the mean of the assigned points. The complexity of this algorithm is O(K * I * N * M), where:

K is the number of clusters
I is the number of iterations
N is the number of data points
M is the number of features
This means that the algorithm takes more time as the number of clusters, iterations, data points, or features increases. However, the number of iterations may vary depending on the initial choice of cluster centers and the convergence criterion. Therefore, the complexity is only an average case estimate.

The testing complexity of KMeans is the time required to assign a new data point to the nearest cluster center. This can be done by computing the distance between the data point and each cluster center, and choosing the minimum. The complexity of this operation is O(K * M), where:

K is the number of clusters
M is the number of features
This means that the algorithm takes more time as the number of clusters or features increases. However, this complexity is much lower than the training complexity, since it does not involve updating the cluster centers or iterating over the entire data set.
